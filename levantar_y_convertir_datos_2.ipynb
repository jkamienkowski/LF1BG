```python
# CASO 3: Todo bien, los separadores de columnas son tabs ('\t') y los de decimales ','
# Pero con una funcion!!!
directorio = 'datos_oscilaciones/'
nombrearchivo='tp4jorig_comma'
data3 = pd.read_csv(directorio + nombrearchivo,header=2,sep='\t')
#data3
```

```python
def convertir_datos(X):
    Y = np.asarray(X)
    
    # Tengo que reemplazar puntos por commas
    i=0
    for line in Y:
        line = line.replace(',', '.')
        Y[i] = float(line)
        i=i+1
    
    return Y
```

```python
t4 = convertir_datos(data3['time'])
x4 = convertir_datos(data3['(N)']) 
```
